## Cleaning df_657

```python
#cdf is the dataframe we are going to clean up
cdf = pd.read_csv('df_657_uncleaned.csv')
cdf.drop(['Unnamed: 0_x', 'Unnamed: 0.1_x', 'Unnamed: 0_y', 'Unnamed: 0.1_y', 'company_name_x'], inplace= True, axis=1)
cdf = cdf.rename(columns={"Unnamed: 0": "Index", "company_name_y": "company_name"})
cdf[cdf['company_name'] == 'AECOM']
```

```python
# some companies, like "AECOM" have mulitple job entries. however the company info is incomplete,
## likely due to a captcha error that was not captured 
# Here I am making sure that all duplicate company_names have the same company info

duplicate_companies = cdf[cdf.duplicated(['company_name'])]
unique_companies = duplicate_companies['company_name'].unique() #list of unique companies
type(duplicate_companies[duplicate_companies['company_name'] == "AECOM"]) #127 duplicate companies
#for each unique value, check the dataframe where company name = unique value

def is_unique(s):
    a = s.to_numpy() # s.values (pandas<0.24)
    return (a[0] == a).all()

def find_value(s):
    for x in s:
        if x != '-1':
            return x

# is_unique(df['counts'])
ci = ['ceo','founded','size','revenue','industry']

for company in unique_companies:
    for index in ci:
        if index != 'founded':
            if not (is_unique(cdf.loc[cdf['company_name'] == company, index])):
                thevalue = find_value(cdf.loc[cdf['company_name'] == company, index])
                (cdf.loc[cdf['company_name'] == company, index]) = (cdf.loc[cdf['company_name'] == company, index]).replace(str(-1), thevalue)
                continue
        if not (is_unique(cdf.loc[cdf['company_name'] == company, index])):
                thevalue = find_value(cdf.loc[cdf['company_name'] == company, index])
                (cdf.loc[cdf['company_name'] == company, index]) = (cdf.loc[cdf['company_name'] == company, index]).replace((-1), thevalue)

## All Duplicate companies have the same company info!
```

```python
#job_title - Title, get rid of parentheses. Sustainability in title, yes or no?

adf['job_title_edit'] = adf['job_title'].apply(lambda x: re.findall("^[^\(]+",x.title())[0])

#Add column identifying if sustainability in title.
adf['sust_title'] = adf['job_title'].apply(lambda x: "Yes" if 'sustainability' in x.lower() else "No")

#adf[adf['sust_title'] == 'Yes']['job_title_edit'] 
#there are 223 job titles with sustainability in them
```

```python
#quick check to see what the common words are in the job titles that have sustainability in them
import spacy

from collections import Counter
nlp = spacy.load('en_core_web_sm')
jobs = ' '.join(adf[adf['sust_title'] == 'Yes']['job_title_edit'].values)
not_in_jobs = ' '.join(adf[adf['sust_title'] == 'No']['job_title_edit'].values)
doc = nlp(jobs)
doc_not = nlp(not_in_jobs)
# all tokens that arent punctuations and not 'sustainability'
words = [token.text
         for token in doc
         if not token.is_punct and token.text.lower() not in ['for', 'energy', 'sustainability', 'and', ' ', 'program', 'environmental', 'solutions', 'global']]

words_not = [token.text
         for token in doc_not
         if not token.is_punct and token.text.lower() not in ['for', 'sustainability', 'and', ' ']]
```

```python
# 15 most common words in jobs without 'sustainability' in the title
word_freq_not = Counter(words_not)
common_words_notitle = word_freq_not.most_common(15)
common_words_notitle
```

```python
# ten most common words in jobs with 'sustainability' in them
word_freq = Counter(words)
common_words_sust = word_freq.most_common(10)
common_words_sust
# common_titles_sust = [x[0] for x in common_words]
## top 10 most common job titles - Manager, Director, Specialist, Engineer, Analyst, Consultant, Coordinator, Intern/Internship, Associate, Senior

# for sj in adf[adf['sust_title'] == 'Yes']['job_title_edit']:
#     if 'manager' in sj:
#         row = 'Manager'
```
